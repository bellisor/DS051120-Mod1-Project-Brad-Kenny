```python
import pandas as pd
import numpy as np
import json
import requests
import config
```

```python
### This loop will iterate through all 500 pages of the TMDB API ###

total_results =[]

### Each iteration it will extract the JSON object from each page and append it into total results ###
for page_num in range(1, 501):
    url = 'https://api.themoviedb.org/3/discover/movie?api_key=' + config.api_key_tmdb + '&language=en-US&region=US&sort_by=original_title.desc&include_adult=false&include_video=false&page=' + str(page_num)
    print("Downloading", url)
    response = requests.get(url)
    data = response.json()
    total_results = total_results + data['results']
    
print("We have", len(total_results), "total results")
```

```python
### Create a data frame ###

df = pd.DataFrame(total_results)
```

```python
df
```

```python

### This gets the links that outputs multiple info from TMDB including IMDB ID ###

links = []
for column in df['id']:
    tmdb_id = str(column)
    links.append('https://api.themoviedb.org/3/movie/' + tmdb_id + '?api_key=' + config.api_key_tmdb + '&language=en-US')
links
```

```python
### Gets the IMDB ID of each movie in the resulting API call above ###

count = 0

for link in links:

    ###  Unique link for each call ###
    response = requests.get(link)
    data = response.json()

    imdb_id = data.get('imdb_id')
    df.at[links.index(link), 'imdb_id'] = imdb_id
```
